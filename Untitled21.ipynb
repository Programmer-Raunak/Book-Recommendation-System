## Popularity Based Recommender System


```python
ratings_with_name = pd.merge(book,rating,on='ISBN', how='inner')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')


avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df

```

```python
new = pd.merge(num_rating_df,avg_rating_df,on='Book-Title', how='inner')
```

```python
new.head()
```

```python
popular_df = new[new['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-L','num_ratings','avg_rating']]
popular_df
```

```python
popular_df.shape
popular_df.columns
```

```python
import pickle
pickle.dump(popular_df,open('top_books.pkl','wb'))
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
experienced_users = x[x].index 
```

```python
experienced_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(experienced_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
pt.index[47]
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    import numpy as np

    # book ka index fetch karo
    index = np.where(pt.index == book_name)[0][0]
    
    # top 5 similar books (excluding itself)
    similar_items = sorted(list(enumerate(similarity_scores[index])),
                           key=lambda x: x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        data.append(book[book['Book-Title'] == pt.index[i[0]]].drop_duplicates('Book-Title')['Book-Title'].values[0])
    
    return data
```

```python
recommend('You Belong To Me')
```

```python
import pickle 
pickle.dump(book,open('book_list.pkl','wb'))
pickle.dump(similarity_scores,open('similarity.pkl','wb'))
pickle.dump(pt,open('pivot_table.pkl','wb'))
```
